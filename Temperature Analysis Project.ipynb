```python
import numpy as np
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Generate random temperature data (°C) for 30 days
temperatures = np.random.randint(20, 40, 30)

# Print the temperatures for each day
print("Daily temperatures (°C):")
print(temperatures)
print()

# 1. Calculate the average temperature
average_temp = np.mean(temperatures)
print(f"1. Average temperature: {average_temp:.2f}°C")

# 2. Find the 5 hottest days (indices)
hottest_indices = np.argsort(temperatures)[-5:]  # Sort indices and get last 5
# Sort them by day number for clarity
hottest_indices = np.sort(hottest_indices)
print(f"2. Indices of 5 hottest days: {hottest_indices}")
print(f"   Corresponding temperatures: {temperatures[hottest_indices]}°C")

# 3. Count days above 35°c
above_35 = np.sum(temperatures > 35)
print(f"3. Days above 85°C: {above_35}")

# 4. Calculate temperature range (max-min)
temp_range = np.max(temperatures) - np.min(temperatures)
print(f"4. Temperature range: {temp_range}°C (from {np.min(temperatures)}°C to {np.max(temperatures)}°C)")

# Bonus: Visualize the temperature data
plt.figure(figsize=(12, 6))
plt.plot(range(1, 31), temperatures, marker='o', linestyle='-', color='red')
plt.axhline(y=average_temp, color='blue', linestyle='--', label=f'Average: {average_temp:.2f}°C')
plt.axhline(y=35, color='green', linestyle=':', label='35°C Threshold')

# Highlight the 5 hottest days
plt.scatter(hottest_indices + 1, temperatures[hottest_indices], 
            color='orange', s=100, zorder=3, label='5 Hottest Days')

plt.xlabel('Day of Month')
plt.ylabel('Temperature (°C)')
plt.title('Daily Temperatures for One Month')
plt.grid(True, alpha=0.3)
plt.legend()
plt.ylim(min(temperatures) - 5, max(temperatures) + 5)
plt.xticks(range(1, 31, 1))  # Show every other day on x-axis for clarity
plt.savefig("Temp.png")
plt.show()
```
